```python
def dataloader(path):
    df = pd.read_csv(path)
    data = df.iloc[:, 1:].values
    data = torch.tensor(data).T
    return data
    
emg_file = 'emg-decomp-fastica-edwin/Data/ramp_10mu_20s/emg_data.csv'
df = pd.read_csv(emg_file)
```

```python
X = torch.tensor([[1.,2.,3.,4.,5.],[11.,12.,13.,14.,15.]]) 
```

```python
D = X - torch.mean(X, -1, True)
D
```

```python
cov = torch.cov(D,correction=0)
cov
```

```python
L, V = torch.linalg.eig(cov)
```

```python
L, V = torch.linalg.eigh(cov,'L')
```

```python
L, V = L.real, V.real
```
